```python
# seleção de todas as linhas onde a idade seja menor que 30
df.loc[df.age <= 30]
```

```python
# seleção de todas as linhas onde a idade seja menor que 35 e mulher
df.loc[(df.age < 35) & (df.sex == 'Female')]
```

```python
# selecionando as 3 primeira linhas
df.loc[1:3]
```

```python
# seleção de algumas informação dentro de um filtro
df.loc[(df.age < 35) & (df.sex == 'Female'), ['age','cp','chol']]
```

```python
# seleção pelo indice 
df.iloc[[0,2,6]]
```

```python
# seleção de indices (0 e 2) e colunas (1 e 3)
df.iloc[[0,2],[1,3]]
```
